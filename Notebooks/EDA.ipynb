```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("employee_data.csv")
df.head()
```

```python
df.info()
```

```python

df.describe()
```

```python

df["Hire_Date"] = pd.to_datetime(df["Hire_Date"], errors='coerce')
df["Resigned"] = df["Resigned"].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Employees with high performance
df[df["Performance_Score"] >= 4]
```

```python
# Employees with salary > 6000
df[df["Monthly_Salary"] > 6000]
```

```python
# High performance AND high salary
df[(df["Performance_Score"] >= 4) & (df["Monthly_Salary"] > 6000)]
```

```python
# High performance AND high salary
df[(df["Performance_Score"] >= 4) & (df["Monthly_Salary"] > 6000)]
```

```python
# IT department OR Finance department
df[(df["Department"] == "IT") | (df["Department"] == "Finance")]
```

```python
# Multiple departments
df[df["Department"].isin(["IT", "Finance", "Engineering"])]
```

```python
# Age between 30 and 50
df[df["Age"].between(30, 50)]
```

```python
# Job titles containing 'Analyst'
df[df["Job_Title"].str.contains("Analyst", case=False)]
```

```python
df.sort_values(by="Performance_Score", ascending=False).head(10)
```

```python
# Lowest performers
df.sort_values(by="Performance_Score").head(5)
```

```python
# Sort by salary (ascending)
df.sort_values(by="Monthly_Salary")

# Descending
df.sort_values(by="Monthly_Salary", ascending=False)
```

```python
# Top 10 highest salary employees
df.sort_values(by="Monthly_Salary", ascending=False).head(10)
```

```python
# Top performing IT employees
df[df["Department"] == "IT"].sort_values(by="Performance_Score", ascending=False)
```

```python
# Average performance per department
dept_perf = df.groupby("Department")["Performance_Score"].mean().sort_values(ascending=False)

print(dept_perf)
```

```python
result = (
    df.sort_values(["Department", "Performance_Score"], ascending=[True, False])
      .groupby("Department")
      .head(1)
)
result
```

```python
# Departments with avg performance > 3
high_dept = df.groupby("Department")["Performance_Score"].mean()
high_dept = high_dept[high_dept > 3]

df[df["Department"].isin(high_dept.index)]
```

```python
sns.histplot(df["Performance_Score"], bins=10)
plt.title("Performance Score Distribution")
plt.show()
```

```python
sns.boxplot(x=df["Monthly_Salary"], color="g")
plt.title("Salary Spread")
plt.show()
```

```python
# Department vs Performance
sns.boxplot(x="Department", y="Performance_Score", data=df , palette="deep")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Gender", data=df , palette= "coolwarm")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Experience vs Performance
sns.scatterplot(x="Years_At_Company", y="Performance_Score", data=df)
plt.show()

# Training vs Performance
sns.scatterplot(x="Training_Hours", y="Performance_Score", data=df)
plt.show()

# Satisfaction vs Performance
sns.scatterplot(x="Employee_Satisfaction_Score", y="Performance_Score", data=df)
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df[[
    "Performance_Score",
    "Training_Hours",
    "Years_At_Company",
    "Monthly_Salary",
    "Employee_Satisfaction_Score"
]])
plt.show()
```

```python
sns.scatterplot(
    x="Training_Hours",
    y="Performance_Score",
    hue="Department",
    data=df
)
plt.title("Training vs Performance")
plt.show()
```

```python
sns.scatterplot(
    x="Employee_Satisfaction_Score",
    y="Performance_Score",
    data=df
)
plt.title("Satisfaction Impact")
plt.show()
```

```python
sns.regplot(
    x="Monthly_Salary",
    y="Performance_Score",
    data=df
)
plt.title("Salary vs Performance (Trend Line)")
plt.show()
```

```python
df["Training_Bins"] = pd.cut(df["Training_Hours"], bins=10)

sns.barplot(
    x="Training_Bins",
    y="Performance_Score",
    data=df
)
plt.xticks(rotation=45)
plt.title("Avg Performance by Training Level")
plt.show()
```

```python
(
    df.groupby("Department")["Performance_Score"]
      .mean()
      .sort_values(ascending=False)
      .plot(kind="bar")
)

plt.title("Department Ranking by Performance")
plt.show()
```
